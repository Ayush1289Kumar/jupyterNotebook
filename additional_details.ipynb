# round()
This function is used to round of a number and also we can provide till how many digits we want to round off.

```python
result = 88/21
print(round(result,1))
print(round(result,2))
print(round(result,3))
print(round(result,4))
print(round(result,5))
```

# Random module 
- random.choice(seq) = random choice from a given sequence.
- random.shuffle(x) = suffles the values of a sequence.
- random.randint(a,b) = both a and b is included.
- random.randrange(a,b) = a is include and b is not included.
- random.random() = a random number between 0 and 1, and 1 is not included.


```python
from random import choice,shuffle,randint,randrange,random

num = [87,34,23,11,990]
print(choice(num))

shuffle(num)
print(num)

i=0
while i!=10:
    print(i)
    i = randint(1,10)

print(round(random(),2))


```

# Command line arguments.

To use command line arguments in python we have to use sys module, which has an attribute sys.argv
where argv is a list contatining the number of arguments passed in the command line while executing the program. 

sys.argv[0] is always the program name. 
sys.exit exits the program. 

```python
import sys

if len(sys.argv) < 2:
    sys.exit('Less arguments')

elif len(sys.argv) > 2:
    sys.exit("Too many argumnets")

else:
    print(f"Hello {sys.argv[1]} how are you.")
```
