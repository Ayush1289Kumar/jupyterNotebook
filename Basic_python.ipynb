# Bitwise Operators

```python
a=2 # 0010
s=8 # 1000

# Bitwise AND : sets each bit to 1 if both are  1
print(f"a & s = {a & s}") # 0010 & 1000 = 0000 => 0
# Bitwise OR : sets each bit to 1 if one of bits is 1
print(f"a | s = {a | s}") # 0010 | 1000 = 1010 => 10
# Bitwise XOR : sets each bit to 1 if either is 1 or are the bits different?
print(f"a ^ s = {a ^ s}") # 0010 ^ 1000 = 1010 => 10
# Bitwise NOT : inverts all the bits
print(f"~a = {~a}") # -a -1 = -2-1 => -3
print(f"~s = {~s}") # -s -1 = -8-1 => -9

# Bitwise Right shift : shifts each bit to the right
print(f"a >> 1 = { a >> 1}") # 0010 >> 1 = 0001 => 1
print(f"s >> 1 = { s >> 1}") # 1000 >> 1 = 0100 => 4
# Bitwise Left shift : shift each bit to the left
print(f"a << 1 = {a << 1}") # 0010 << 1 = 0100 => 4
print(f"s << 1 = {s << 1}") # 1000 << 1 = 10000 => 16 
```

# Set Special Operations
- union() =  returns a new set with al items from both the sets.
- update() = inserts all items from one set into another.
- copy() = creates a copy of the set.
- intersection() = returns a new set with only the duplicate values.
- intersection_update() = updates the value to intersection values.
- difference() = returns a new set that will contain only the items from the first set and that are not  
present in the other set.
- difference_update() = updates the value of the first set with the difference
- symmetric_difference() = returns a set which keeps only the values that are not present in both sets.
- symmetric difference_update() = updates the set with symmetric different values.
- isdisjoint() = returns whether two sets have a intersection or not.
- issubset() = returns whether another set contains this set or not.
- issuperset() = returns whether this set contains another set or not.



```python
# union() 
set1= {"Eren","Mikasa","Hinata","Naruto","Kira"}
set2 = {"Bleach","One Piece","Naruto"}
set3={"Attack on titan","Levi","Reiner","Eren","Annie","Mikasa"}
new1 = set1.union(set2)
print(new1)

# multiple sets
new2 = set1.union(set2,set3)
print(new2)

# Union using  | operator
new3 = set1 | set3
print(new3)

# Multiple sets using | operator
new4= set1|set3|set2
print(new4)

# joining sets elements with different iterables.
new5=set1.union(list(["jjk","opm"]))
print(new5)
```

```python
# update()
set1 = {"Eren", "Mikasa", "Hinata", "Naruto", "Kira"}
set2 = {"Bleach", "One Piece", "Naruto"}
set3 = {"Attack on titan","Levi","Reiner","Eren","Annie","Mikasa"}
set1.update(set2)
print(set1)
```

```python
# intesection() 
set1 = {"Eren", "Mikasa", "Hinata", "Naruto", "Kira"}
set2 = {"Bleach", "One Piece", "Naruto"}
set3 = {"Attack on titan","Levi","Reiner","Eren","Annie","Mikasa"}

new1 = set1.intersection(set3)
print(new1)

# intersection_update()
set1.intersection_update(set2)
print(set1)

# using & operator
new2 = set1 & set2
print(new2)
```

```python
# difference()
set1 = {"Eren", "Mikasa", "Hinata", "Naruto", "Kira"}
set2 = {"Bleach", "One Piece", "Naruto"}
set3 = {"Attack on titan","Levi","Reiner","Eren","Annie","Mikasa"}

new = set1.difference(set2)
print(new)

# using difference_update()
set1.difference_update(set3)
print(set1)

# using - operator
new = set1 -set2
print(new)
```

```python
# using symmetric_difference()
set1 = {"Eren", "Mikasa", "Hinata", "Naruto", "Kira"}
set2 = {"Bleach", "One Piece", "Naruto"}
set3 = {"Attack on titan","Levi","Reiner","Eren","Annie","Mikasa"}

new = set1.symmetric_difference(set2)
print(new)

# using symmetric_difference_update()
set2.symmetric_difference_update(set1)
print(set2)

# using ^ operator
new = set2^set3
print(new)

```

```python
# checking if disjoint set or not.
set1 = {"Eren", "Mikasa", "Hinata", "Naruto", "Kira"}
set2 = {"Bleach", "One Piece", "Naruto"}
set3 = {"Attack on titan", "Levi", "Reiner", "Eren", "Annie", "Mikasa"}

print(set1.isdisjoint(set2)) # False
print(set2.isdisjoint(set3)) # True

# checking if subset or not.
print(set1.issubset(set3)) # False
set1.discard("Hinata")
set1.discard("Naruto")
set1.discard("Kira")
print(set1.issubset(set3)) # True

# checking if superset of not.
print(set3.issuperset(set1)) # True

set1.add("Kira")

print(set3.issuperset(set1)) # False
```

## Number of argumnets
- Arbitrary arguments= def func(*args)  recieves a tuple.
- Arbitrary keyword arguments = def func(**kwargs) recieves a dictionary.
- default argument =  def func(arg=default value)

```python
# arbitrary arguments.

def f(*args):
    if "Uchiha" not in args[0]:
        print("Are u a fool!!!")
    else:
        print(f"{args[0]} = ðŸ’ª")

f("Hashirama","Obito Uchiha")

f("Madara Uchiha","Naruto")

```

```python
# arbitrary kwyword arguments.
def f(**kwargs):
    if kwargs["special"] == "Sharingan":
        print(f"{kwargs["name"]} = ðŸ’ª")
    else:
        print("Weak!!!!!!")

f(name="Kakashi",special="Sharingan")
f(name="Shikamaru",special="Intelligence")

```

# Positional-only arguments.
To specify that a function can have only postional arguments
        add , / after the arguments


```python
# Postional only arguments.
def fun(x,y,/):
    print(f"{x} and {y}")

fun(123,322)

# fun(y=123,x=322)    will raise TypeError
```

# Keyword-only arguments.
To specify that a function have only keyword arguments add *, before the arguments.

```python
# Keyword only arguments.
def fun(*,x,y):
    print(f"{y} is good and {x} is bad")

fun(x="Kakashi",y="Obito")

# fun("Kakashi","Obito") will raise TypeError
```

# Lambda function 
A lambda function is a small annonymous function which can take any number of arguments,

but can only have one expression,

the expression is executed and the result is returned.

- Use lambda functions when an anonymous function is required for a short period of time.

## Syntax



```python
# variable = lambda arg : expression

square = lambda n: n**2

print(square(9))
```
