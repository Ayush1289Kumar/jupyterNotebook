## Shortcuts:
- ctrl + enter = run
- shift + enter = run and move to new cell
- y = code mode
- m = markdown cell
- dd = delete 
- Arrow keys to navigate  or using j and k
- ctrl + arrow key = move cell 
- a = insert cell above
- b = insert cell below
- x = cut
- c = copy
- v = paste
- z = undo delete
