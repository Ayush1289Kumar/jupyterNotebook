# JSON ( JavaScript Object Notation)
It is a syntax fro storing and exchanging data


Python has a built-in package calle json, which can be used to work with JSON data.

## Parse json 
Convert from Json to Python
If we have a Json string. we can parse it by using the json.loads() method

### __Note__: The result will be a python dictionary.


```python
import json

# opening json file and storing its data
with open("AttackOnTitan.json","r") as file:
    data = file.read()

new_data = json.loads(data)
print(new_data["anime"])
```

# Convert from python to json
If we have a python object we can convert it into a json string by using the json.dumps() method

```python
import json
new = {
    'name' : "Eren yeager",
    'titan' : "Attack Titan",
}

new_ = json.dumps(new)
print(new_)
```

# Python objects into Json
When we convert from python to Json, Python objects are converted into the Json equivalent.
|    Python   | Json  |
|-------------|-------|
| dict        | object|
| list        | Array |
| tuple       | Array |
| str         | String|
| int         | Number|
| float       | Number|
| True,False  | true,false|
| None        | null |


```python
new = {
    "name":"Naruto uzumaki",
    "age" : 32,
    "married" : True,
    "divorced" : False,
    "child" : ("Boruto,Himawari"),
    "pets" : None,
    "strengths" : ["Talk no jutsu","Never giving up"]
}

new_ = json.dumps(new)
print(new_)
```

# Format the result
- The example above prints a Json string but it is not very easy to read with no indentation and line breaks.
- The json.dumps() method has parameter to make it easier to read the result.

```python
# using indent= parameter to define the number of indents.

print(json.dumps(new,indent=1))
# the more is the value of indent the more will be the indentation & spaces.
```

We can also define the seperators,

default value is (", ",": ") which means using a comma to seperate each object and a space to seperate keys from values.

```python
# using the separators= parameter to change the default separator.

print(json.dumps(new,indent=1,separators=(". ","=> ")))
```

# Order the result.
The json.dumps() method has a parameter to order the keys in the result.


```python
# using the sort_keys= Parameter to specify if the result should be shorted or not.

print(json.dumps(new,indent=1,sort_keys=True))
```
